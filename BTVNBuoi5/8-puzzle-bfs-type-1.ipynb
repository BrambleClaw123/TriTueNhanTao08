```python
class Node :
    def __init__(self, state, parent, action, cost):
        self.state = state
        self.parent = parent
        self.action = action
        self.cost = cost
    def print_node(self):
        print(f"State: {self.state}")
        print(f"Parent: {self.parent}")
        print(f"Action: {self.action}")
        print(f"Cost: {self.cost}")
        print()
a = Node([[1, 2, 3],[4, 5, 6], [7,8,0]], "", "", 0)
print(a.state)
```

```python
class Problem:
    def __init__(self, initial, goal):
        self.initial = initial
        self.goal = goal
    def is_goal(self, state):
        for i in range(len(state)):
            for j in range(len(state[0])):
                if self.goal[i][j] != state[i][j]:
                    return False
        return True
problem = Problem([[1, 9, 8],[0, 5, 6], [7,2,3]], [[1, 2, 3],[4, 5, 6], [7,8,0]])
print(problem.is_goal(a.state))
```

```python
class Frontier:
    def __init__(self):
        self.queue = []
    def enqueue(self, node):
        self.queue.append(node)
    def dequeue(self):
        return self.queue.pop(0)
    def is_empty(self):
        return True if len(self.queue) == 0 else False
    def is_contain(self, node):
        return True if node in self.queue else False
frontier = Frontier()
frontier.enqueue(a)
print(frontier.queue[0].state)
print(frontier.is_empty())
print(frontier.is_contain(a))
b = frontier.dequeue()
print(b.state)
print(frontier.is_empty())
```

```python
import copy
def expand(node):
    state = node.state
    action = []
    for i in range(len(state)):
        for j in range(len(state[0])):
            if (state[i][j] == 0):
                loc = (i, j)
                if (i!=0):
                    action.append((-1, 0))
                if (i!=len(state)-1):
                    action.append((1, 0))
                if (j!=0):
                    action.append((0, -1))
                if (j!= len(state[0])-1):
                    action.append((0, 1))
    child = []
    for act in action:
        clone_state = copy.deepcopy(state)
        x1 = loc[0]
        y1 = loc[1]
        x2 = loc[0] + act[0]
        y2 = loc[1] + act[1]
        clone_state[x1][y1], clone_state[x2][y2] = clone_state[x2][y2], clone_state[x1][y1]
        new_node = copy.deepcopy(node)
        new_node.state = clone_state
        new_node.parent = node
        new_node.cost = node.cost + 1
        if (act[0] == 1):
            new_node.action = "DOWN"
        if (act[0] == -1):
            new_node.action = "UP"
        if (act[1] == -1):
            new_node.action = "LEFT"
        if (act[1] == 1):
            new_node.action = "RIGHT"
        child.append(new_node)
    return child
for child in expand(a):
    child.print_node()
```

```python
def BFS(problem):
    node = Node(problem.initial, "", "", 0)
    frontier = Frontier()
    frontier.enqueue(node)
    reached = [node.state]
    while frontier.is_empty() == False:
        node = frontier.dequeue()
        if problem.is_goal(node.state) == True:
            return node
        for child in expand(node):
            state = child.state
            if state not in reached and frontier.is_contain(child) == False:
                reached.append(state)
                frontier.enqueue(child)
    failure = "Khong the giai duoc"
    return failure
    
```

```python
import tkinter as tk
import time

def perform(state, cells_state):
    for i in range(len(state)):
        for j in range(len(state[0])):
            val = state[i][j]
            if val == 0:
                cells_state[i][j].config(text="", bg="#e6e6e6", relief="solid", bd=1)
            else:
                cells_state[i][j].config(text=str(val), bg="#b3b3b3", fg="black", relief="solid", bd=2)

def write_to_log(txt_log, message):
    txt_log.config(state=tk.NORMAL)       
    txt_log.insert(tk.END, message + "\n") 
    txt_log.config(state=tk.DISABLED)     
    txt_log.see(tk.END)                   

def handle_click(lbl_status, btn_run, problem, root, cells_state, txt_log):
    lbl_status.config(text="Đang tính toán ...", fg="blue")
    btn_run.config(state=tk.DISABLED)
    
    txt_log.config(state=tk.NORMAL)
    txt_log.delete("1.0", tk.END)
    txt_log.config(state=tk.DISABLED)
    root.update()
    
    result = BFS(problem)
        
    if type(result) is str:
        lbl_status.config(text=result, fg="red")
        write_to_log(txt_log, f"Lỗi: {result}")
        btn_run.config(state=tk.NORMAL)
        root.update()
        return
        
    current = result
    path = []
    while current != "":
        path.append(current)
        current = current.parent
    path.reverse()

    for i, node in enumerate(path):
        perform(node.state, cells_state)
        if i == 0:
            mess = "Trạng thái ban đầu"
        else:
            mess = f"Bước {i}: Hành động thực hiện [{node.action}]"
            
        lbl_status.config(text=mess, fg="black")
        write_to_log(txt_log, mess)
        root.update()
        time.sleep(1.25)
        
    final_mess = f"Đã xong, tổng số bước là {len(path)-1}"
    lbl_status.config(text=final_mess, fg="green")
    write_to_log(txt_log, f"\n========\n{final_mess}")
    
    problem.initial = result.state
    btn_run.config(state=tk.NORMAL)

def main():
    initial = [[8, 2, 3],
               [1, 6, 4],
               [7, 0, 5]]
    
    goal    = [[1, 2, 3],
               [4, 5, 6],
               [7, 8, 0]]
    
    problem = Problem(initial, goal)
    cells_state = [[None]*3 for _ in range(3)]
    
    root = tk.Tk()
    root.title("8 Puzzle BFS Type 1")
    root.configure(bg="#f0f0f0")
    
    top_panel = tk.Frame(root, bg="#f0f0f0", padx=10, pady=10, bd=1, relief="solid")
    top_panel.grid(row=0, column=0, columnspan=2, sticky="ew")
    
    btn_run = tk.Button(top_panel, text="Play (Giải)", font=("Arial", 10, "bold"), bg="#ffffff", fg="green")
    btn_run.pack(side=tk.LEFT, padx=5)
    
    matrix_panel = tk.Frame(root, bg="#f0f0f0", padx=20, pady=20)
    matrix_panel.grid(row=1, column=0, sticky="n")
    
    for i in range(3):
        for j in range(3):
            lbl = tk.Label(matrix_panel, text="", font=("Arial", 22, "bold"), width=5, height=2)
            lbl.grid(row=i, column=j, padx=4, pady=4)
            cells_state[i][j] = lbl
            
    perform(problem.initial, cells_state)

    lbl_status = tk.Label(matrix_panel, text="Sẵn sàng chạy mô phỏng", font=("Arial", 10, "italic"), bg="#f0f0f0")
    lbl_status.grid(row=3, column=0, columnspan=3, pady=(15, 0))

    log_panel = tk.Frame(root, bg="#f0f0f0", padx=15, pady=20)
    log_panel.grid(row=1, column=1, sticky="n")

    txt_log = tk.Text(log_panel, width=38, height=16, font=("Consolas", 10), 
                      bg="#ffffff", fg="#333333", relief="solid", bd=1, padx=5, pady=5)
    txt_log.grid(row=0, column=0, padx=5, pady=5, sticky="nsew")
    txt_log.config(state=tk.DISABLED)

    btn_run.config(command=lambda: handle_click(lbl_status, btn_run, problem, root, cells_state, txt_log))

    root.mainloop()

main()
```
