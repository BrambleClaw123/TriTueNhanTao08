```python
class Node :
    def __init__(self, state, parent, action, cost):
        self.state = state
        self.parent = parent
        self.action = action
        self.cost = cost
    def __lt__(self, other):
        return self.cost < other.cost
    def print_node(self):
        print(f"State: {self.state}")
        print(f"Parent: {self.parent}")
        print(f"Action: {self.action}")
        print(f"Cost: {self.cost}")
        print()
a = Node([[1, 2, 3],[4, 5, 6], [7,8,0]], "", "", 0)
print(a.state)
```

```python
class Problem:
    def __init__(self, initial, goal):
        self.initial = initial
        self.goal = goal
    def is_goal(self, state):
        for i in range(len(state)):
            for j in range(len(state[0])):
                if self.goal[i][j] != state[i][j]:
                    return False
        return True
problem = Problem([[1, 9, 8],[0, 5, 6], [7,2,3]], [[1, 2, 3],[4, 5, 6], [7,8,0]])
print(problem.is_goal(a.state))
```

```python
class Frontier:
    def __init__(self):
        self.queue = []
    def enqueue(self, node):
        self.queue.append(node)
    def dequeue(self):
        return self.queue.pop(0)
    def is_empty(self):
        return True if len(self.queue) == 0 else False
    def is_contain(self, node):
        return True if node in self.queue else False
frontier = Frontier()
frontier.enqueue(a)
print(frontier.queue[0].state)
print(frontier.is_empty())
print(frontier.is_contain(a))
b = frontier.dequeue()
print(b.state)
print(frontier.is_empty())
```

```python
import copy
def expand(node):
    state = node.state
    action = []
    for i in range(len(state)):
        for j in range(len(state[0])):
            if (state[i][j] == 0):
                loc = (i, j)
                if (j!=0):
                    action.append((0, -1))
                if (j!= len(state[0])-1):
                    action.append((0, 1))
                if (i!=0):
                    action.append((-1, 0))
                if (i!=len(state)-1):
                    action.append((1, 0))
    child = []
    for act in action:
        clone_state = copy.deepcopy(state)
        x1 = loc[0]
        y1 = loc[1]
        x2 = loc[0] + act[0]
        y2 = loc[1] + act[1]
        clone_state[x1][y1], clone_state[x2][y2] = clone_state[x2][y2], clone_state[x1][y1]
        new_node = Node(clone_state, node, "", node.cost + 1)
        if (act[0] == 1):
            new_node.action = "DOWN"
        if (act[0] == -1):
            new_node.action = "UP"
        if (act[1] == -1):
            new_node.action = "LEFT"
        if (act[1] == 1):
            new_node.action = "RIGHT"
        child.append(new_node)
    return child
for child in expand(a):
    child.print_node()
```

```python
def BFS_type_1(problem):
    node = Node(problem.initial, "", "", 0)
    frontier = Frontier()
    frontier.enqueue(node)
    reached = [node.state]
    while frontier.is_empty() == False:
        node = frontier.dequeue()
        if problem.is_goal(node.state) == True:
            return node
        for child in expand(node):
            state = child.state
            if state not in reached and frontier.is_contain(child) == False:
                reached.append(state)
                frontier.enqueue(child)
    failure = "Không thể giải được"
    return failure
    
```

```python
def BFS_type_2(problem):
    node = Node(problem.initial, "", "", 0)
    if problem.is_goal(node.state) == True:
        return node
    frontier = Frontier()
    frontier.enqueue(node)
    reached = [node.state]
    while frontier.is_empty() != True:
        node = frontier.dequeue()
        for child in expand(node):
            state = child.state
            if problem.is_goal(state) == True:
                return child
            if state not in reached and frontier.is_contain(child) == False:
                reached.append(state)
                frontier.enqueue(child)
    failure = "Không thể giải được"
    return failure
```

```python
def DFS(problem):
    node = Node(problem.initial, "", "", 0)
    if problem.is_goal(node.state) == True:
        return node
    frontier = Frontier_2()
    frontier.push(node)
    reached = [node.state]
    while frontier.is_empty() == False:
        node = frontier.pop()
        for child in expand(node):
            state = child.state
            if problem.is_goal(node.state) == True:
                return node
            if state not in reached and frontier.is_contain(child) == False:
                reached.append(state)
                frontier.push(child)
    failure = "Không thể giải được"
    return failure
    
```

```python
class Frontier_2:
    def __init__(self):
        self.queue = []
    def push(self, node):
        self.queue.append(node)
    def pop(self):
        return self.queue.pop(len(self.queue) - 1)
    def is_empty(self):
        return True if len(self.queue) == 0 else False
    def is_contain(self, node):
        return True if node in self.queue else False
frontier = Frontier()
frontier.enqueue(a)
print(frontier.queue[0].state)
print(frontier.is_empty())
print(frontier.is_contain(a))
b = frontier.dequeue()
print(b.state)
print(frontier.is_empty())
```

```python
def IDS(problem):
    def DLS(problem, depth):
        node = Node(problem.initial, "", "", 0)
        if problem.is_goal(node.state) == True:
            return node
        result = "Không thể giải được"
        frontier = Frontier_2()
        frontier.push(node)
        reached = [node.state]
        while frontier.is_empty() != True:
            node = frontier.pop()
            if node.cost >= depth:
                result = "cutoff"
                continue
            for child in expand(node):
                state = child.state
                if problem.is_goal(state):
                    return child
                if state not in reached and not frontier.is_contain(child):
                    reached.append(state)
                    frontier.push(child)
        return result
            
    depth = 0
    while True:
        result = DLS(problem, depth)
        if result != "cutoff":
            return result
        depth += 1
```

```python
def count_wrong(state):
    mat = [[1, 2, 3],
           [4, 5, 6],
           [7, 8, 0]]
    cost = 0
    for i in range(3):
        for j in range(3):
            if mat[i][j] != state[i][j]:
                cost += 1
    return cost
```

```python
import copy
def expand_UCS(node):
    state = node.state
    action = []
    for i in range(len(state)):
        for j in range(len(state[0])):
            if (state[i][j] == 0):
                loc = (i, j)
                if (j!=0):
                    action.append((0, -1))
                if (j!= len(state[0])-1):
                    action.append((0, 1))
                if (i!=0):
                    action.append((-1, 0))
                if (i!=len(state)-1):
                    action.append((1, 0))
    child = []
    for act in action:
        clone_state = copy.deepcopy(state)
        x1 = loc[0]
        y1 = loc[1]
        x2 = loc[0] + act[0]
        y2 = loc[1] + act[1]
        clone_state[x1][y1], clone_state[x2][y2] = clone_state[x2][y2], clone_state[x1][y1]
        new_node = Node(clone_state, node, "", node.cost + count_wrong(clone_state))
        if (act[0] == 1):
            new_node.action = "DOWN"
        if (act[0] == -1):
            new_node.action = "UP"
        if (act[1] == -1):
            new_node.action = "LEFT"
        if (act[1] == 1):
            new_node.action = "RIGHT"
        child.append(new_node)
    return child
```

```python
import heapq
def UCS(problem):
    node = Node(problem.initial, "", "", count_wrong(problem.initial))
    frontier = []
    heapq.heappush(frontier, node)
    reached = {}
    while len(frontier) != 0:
        node = heapq.heappop(frontier)
        if node.cost > reached.get(str(node.state),float('inf')):
            continue
        reached[str(node.state)] = node.cost
        if problem.is_goal(node.state) == True:
            return node
        for child in expand_UCS(node):
            state = child.state
            if str(state) not in reached or child.cost < reached.get(str(state),float('inf')):
                heapq.heappush(frontier, child)
    failure = "Không thể giải được"
    return failure
```

```python
def manhattan(state):
    goal = [[1, 2, 3],
            [4, 5, 6],
            [7, 8, 0]]
    cost = 0
    for i in range(3):
        for j in range(3):
            if state[i][j] == 1:
                cost += abs(i - 0) + abs(j - 0)
            elif state[i][j] == 2:
                cost += abs(i - 0) + abs(j - 1)
            elif state[i][j] == 3:
                cost += abs(i - 0) + abs(j - 2)
            elif state[i][j] == 4:
                cost += abs(i - 1) + abs(j - 0)
            elif state[i][j] == 5:
                cost += abs(i - 1) + abs(j - 1)
            elif state[i][j] == 6:
                cost += abs(i - 1) + abs(j - 2)
            elif state[i][j] == 7:
                cost += abs(i - 2) + abs(j - 0)
            elif state[i][j] == 8:
                cost += abs(i - 2) + abs(j - 1)
    return cost
print(manhattan([[1, 2, 3],[4, 0, 6], [7, 5, 8]]))
```

```python
def GSA(problem):
    node = Node(problem.initial, "", "", manhattan(problem.initial))
    frontier = []
    heapq.heappush(frontier, node)
    reached = {}
    while len(frontier) != 0:
        node = heapq.heappop(frontier)
        if node.cost > reached.get(str(node.state), float('inf')):
            continue
        reached[str(node.state)] = node.cost
        if problem.is_goal(node.state) == True:
            return node
        for child in expand_UCS(node):
            state = child.state
            if str(state) not in reached or child.cost < reached[str(state)]:
                heapq.heappush(frontier, child)
    failure = "Không thể giải được"
    return failure
```

```python
import copy
def expand_A_star(node):
    state = node.state
    action = []
    for i in range(len(state)):
        for j in range(len(state[0])):
            if (state[i][j] == 0):
                loc = (i, j)
                if (j!=0):
                    action.append((0, -1))
                if (j!= len(state[0])-1):
                    action.append((0, 1))
                if (i!=0):
                    action.append((-1, 0))
                if (i!=len(state)-1):
                    action.append((1, 0))
    child = []
    for act in action:
        clone_state = copy.deepcopy(state)
        x1 = loc[0]
        y1 = loc[1]
        x2 = loc[0] + act[0]
        y2 = loc[1] + act[1]
        clone_state[x1][y1], clone_state[x2][y2] = clone_state[x2][y2], clone_state[x1][y1]
        new_node = Node(clone_state, node, "", f(clone_state, node))
        if (act[0] == 1):
            new_node.action = "DOWN"
        if (act[0] == -1):
            new_node.action = "UP"
        if (act[1] == -1):
            new_node.action = "LEFT"
        if (act[1] == 1):
            new_node.action = "RIGHT"
        child.append(new_node)
    return child
```

```python
def A_star(problem):
    node = Node(problem.initial, "", "", manhattan(problem.initial) + count_wrong(problem.initial))
    frontier = []
    heapq.heappush(frontier, node)
    reached = {}
    while len(frontier) != 0:
        node = heapq.heappop(frontier)
        if node.cost > reached.get(str(node.state), float('inf')):
            continue
        reached[str(node.state)] = node.cost
        if problem.is_goal(node.state) == True:
            return node
        for child in expand_A_star(node):
            state = child.state
            if str(state) not in reached or child.cost < reached[str(state)]:
                heapq.heappush(frontier, child)
    failure = "Không thể giải được"
    return failure
```

```python
def IDA_star(problem):
    def DLS(problem, cost):
        node = Node(problem.initial, "", "", manhattan(problem.initial) + count_wrong(problem.initial))
        result = "Không thể giải được"
        frontier = []
        heapq.heappush(frontier, node)
        reached = {}
        while len(frontier) != 0:
            node = heapq.heappop(frontier)
            if node.cost > reached.get(str(node.state), float('inf')):
                continue
            reached[str(node.state)] = node.cost
            if problem.is_goal(node.state):
                return node
            if node.cost >= cost:
                result = "cutoff"
                continue
            for child in expand_A_star(node):
                state = child.state
                if str(state) not in reached or child.cost < reached[str(state)]:
                    heapq.heappush(frontier, child)
        return result
            
    cost = manhattan(problem.initial) + count_wrong(problem.initial)
    while True:
        result = DLS(problem, cost)
        if result != "cutoff":
            return result
        cost += 1
```

```python
import tkinter as tk
from tkinter import ttk
import time
def perform(state, cells_state):
    for i in range(len(state)):
        for j in range(len(state[0])):
            val = state[i][j]
            if val == 0:
                cells_state[i][j].config(text="", bg="#D9C3B0", relief="sunken", bd=3)
            else:
                cells_state[i][j].config(
                    text=str(val), 
                    bg="#F5E6D3", 
                    fg="#4A2E1B", 
                    relief="raised", 
                    bd=4
                )

def write_to_log(txt_log, message):
    txt_log.config(state=tk.NORMAL)       
    txt_log.insert(tk.END, message + "\n") 
    txt_log.config(state=tk.DISABLED)     
    txt_log.see(tk.END)                   

def handle_click(lbl_status, btn_run, problem, root, cells_state, txt_log, selected_algo):
    lbl_status.config(text="Đang tính toán ...", fg="#8B4513") 
    btn_run.config(state=tk.DISABLED)
    txt_log.config(state=tk.NORMAL)
    txt_log.delete("1.0", tk.END)
    txt_log.config(state=tk.DISABLED)
    root.update()
    algo_name = selected_algo.get()
    write_to_log(txt_log, f"Đang chạy thuật toán {algo_name}")
    root.update()
    if algo_name == "BFS Type 1":
        result = BFS_type_1(problem)
    elif algo_name == "BFS Type 2":
        result = BFS_type_2(problem)
    elif algo_name == "DFS":
        result = DFS(problem)
    elif algo_name == "IDS":
        result = IDS(problem)
    elif algo_name == "UCS":
        result = UCS(problem)
    elif algo_name == "GSA":
        result = GSA(problem)
    elif algo_name == "A*":
        result = A_star(problem)
    elif algo_name == "IDA*":
        result = IDA_star(problem)
        
    if type(result) is str:
        lbl_status.config(text=result, fg="#A52A2A")
        write_to_log(txt_log, f"Lỗi: {result}")
        btn_run.config(state=tk.NORMAL)
        root.update()
        return
        
    current = result
    path = []
    while current != "":
        path.append(current)
        current = current.parent
    path.reverse()

    for i, node in enumerate(path):
        perform(node.state, cells_state)
        if i == 0:
            mess = "Trạng thái ban đầu"
        else:
            mess = f"Bước {i}: Hành động thực hiện [{node.action}]"
            lbl_status.config(text=mess, fg="#4A2E1B")
            mess += "\nTrạng thái: \n"
            for row in node.state:
                for value in row:
                    mess += str(value) + " "
                mess += "\n"
            mess += "\n"
            
        write_to_log(txt_log, mess)
        root.update()
        time.sleep(0.75)
        
    final_mess = f"Đã xong, tổng số bước là {len(path)-1}"
    write_to_log(txt_log, f"\n========\n{final_mess}")

    final_mess += "\nCác bước đã thực hiện\n" + " -> ".join(node.action for node in path)[4::]
    lbl_status.config(text=final_mess, fg="#2E5A1C", wraplength=320)
    
    problem.initial = result.state
    btn_run.config(state=tk.NORMAL)

def handle_reset(problem, cells_state, lbl_status, txt_log, original_initial):
    problem.initial = [row[:] for row in original_initial]
    perform(problem.initial, cells_state)
    
    txt_log.config(state=tk.NORMAL)
    txt_log.delete("1.0", tk.END)
    txt_log.config(state=tk.DISABLED)
    
    lbl_status.config(text="Sẵn sàng chạy mô phỏng", fg="#4A2E1B")

def main():
    initial = [[0, 4, 3],
               [2, 5, 6],
               [1, 7, 8]]
    
    goal    = [[1, 2, 3],
               [4, 5, 6],
               [7, 8, 0]]
    
    problem = Problem(initial, goal)
    cells_state = [[None]*3 for _ in range(3)]
    
    root = tk.Tk()
    root.title("8 Puzzle")
    
    BG_MAIN = "#EFE5D3"      
    BG_PANEL = "#E6D5C3"     
    COLOR_TEXT = "#4A2E1B"   
    
    root.configure(bg=BG_MAIN)
    root.resizable(False, False) 
    
    style = ttk.Style()
    style.theme_use('clam')
    style.configure("TCombobox", 
                    fieldbackground="#F5E6D3", 
                    background=BG_PANEL, 
                    foreground=COLOR_TEXT, 
                    darkcolor=COLOR_TEXT, 
                    lightcolor="#F5E6D3")
    
    top_panel = tk.Frame(root, bg=BG_PANEL, padx=15, pady=12, bd=2, relief="groove")
    top_panel.grid(row=0, column=0, columnspan=2, sticky="ew", padx=15, pady=(15, 5))

    btn_run = tk.Button(top_panel, text="PLAY", font=("Times New Roman", 10, "bold"), 
                        bg="#F5E6D3", fg="#2E5A1C", activebackground="#E6D5C3",
                        relief="raised", bd=3, width=8, pady=2)  
    btn_run.pack(side=tk.LEFT, padx=5)

    btn_reset = tk.Button(top_panel, text="RESET", font=("Times New Roman", 10, "bold"), 
                          bg="#F5E6D3", fg="#A52A2A", activebackground="#E6D5C3",
                          relief="raised", bd=3, width=8, pady=2) 
    btn_reset.pack(side=tk.LEFT, padx=5)

    lbl_algo = tk.Label(top_panel, text="Thuật toán:", font=("Times New Roman", 10, "italic"), 
                        bg=BG_PANEL, fg=COLOR_TEXT)
    lbl_algo.pack(side=tk.LEFT, padx=(25, 5))
    
    selected_algo = tk.StringVar(value="BFS Type 1")
    root.option_add('*TCombobox*Listbox.font', ("Times New Roman", 10, "bold"))
    algo_combobox = ttk.Combobox(top_panel, textvariable=selected_algo, state="readonly", width=12, font=("Times New Roman", 10, "bold"))
    algo_combobox['values'] = ("BFS Type 1", "BFS Type 2", "DFS", "IDS", "UCS", "GSA", "A*", "IDA*")
    algo_combobox.pack(side=tk.LEFT, padx=5)
    
    matrix_panel = tk.Frame(root, bg="#D9C3B0", padx=15, pady=15, bd=3, relief="sunken")
    matrix_panel.grid(row=1, column=0, padx=(15, 5), pady=(5, 15), sticky="n")
    
    for i in range(3):
        for j in range(3):
            lbl = tk.Label(matrix_panel, text="", font=("Times New Roman", 24, "bold"), width=4, height=2)
            lbl.grid(row=i, column=j, padx=6, pady=6)
            cells_state[i][j] = lbl
            
    perform(problem.initial, cells_state)

    lbl_status = tk.Label(root, text="Sẵn sàng chạy mô phỏng", font=("Times New Roman", 11, "italic"), 
                          bg=BG_MAIN, fg=COLOR_TEXT, justify=tk.CENTER)
    lbl_status.grid(row=2, column=0, columnspan=2, pady=(0, 15), padx=15, sticky="ew")

    log_panel = tk.Frame(root, bg=BG_MAIN)
    log_panel.grid(row=1, column=1, padx=(5, 15), pady=(5, 5), sticky="nsew")

    txt_log = tk.Text(log_panel, width=38, height=15, font=("Times New Roman", 10, "bold"), 
                      bg="#FDFBF7", fg="#3E2723", relief="sunken", bd=2, padx=8, pady=8)
    txt_log.grid(row=0, column=0, sticky="nsew")
    log_panel.grid_rowconfigure(0, weight=1)
    log_panel.grid_columnconfigure(0, weight=1)
    txt_log.config(state=tk.DISABLED)

    btn_run.config(command=lambda: handle_click(lbl_status, btn_run, problem, root, cells_state, txt_log, selected_algo))
    btn_reset.config(command=lambda: handle_reset(problem, cells_state, lbl_status, txt_log, initial))

    root.mainloop()

main()
```
