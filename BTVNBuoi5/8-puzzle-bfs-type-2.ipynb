```python
class Node :
    def __init__(self, state, parent, action, cost):
        self.state = state
        self.parent = parent
        self.action = action
        self.cost = cost
    def print_node(self):
        print(f"State: {self.state}")
        print(f"Parent: {self.parent}")
        print(f"Action: {self.action}")
        print(f"Cost: {self.cost}")
        print()
a = Node([[1, 2, 3],[4, 5, 6], [7,8,0]], "", "", 0)
print(a.state)
```

```python
class Problem:
    def __init__(self, initial, goal):
        self.initial = initial
        self.goal = goal
    def is_goal(self, state):
        for i in range(len(state)):
            for j in range(len(state[0])):
                if self.goal[i][j] != state[i][j]:
                    return False
        return True
problem = Problem([[1, 9, 8],[0, 5, 6], [7,2,3]], [[1, 2, 3],[4, 5, 6], [7,8,0]])
print(problem.is_goal(a.state))
```

```python
class Frontier:
    def __init__(self):
        self.queue = []
    def enqueue(self, node):
        self.queue.append(node)
    def dequeue(self):
        return self.queue.pop(0)
    def is_empty(self):
        return True if len(self.queue) == 0 else False
    def is_contain(self, node):
        return True if node in self.queue else False
frontier = Frontier()
frontier.enqueue(a)
print(frontier.queue[0].state)
print(frontier.is_empty())
print(frontier.is_contain(a))
b = frontier.dequeue()
print(b.state)
print(frontier.is_empty())
```

```python
import copy
def expand(node):
    state = node.state
    action = []
    for i in range(len(state)):
        for j in range(len(state[0])):
            if (state[i][j] == 0):
                loc = (i, j)
                if (i!=0):
                    action.append((-1, 0))
                if (i!=len(state)-1):
                    action.append((1, 0))
                if (j!=0):
                    action.append((0, -1))
                if (j!= len(state[0])-1):
                    action.append((0, 1))
    child = []
    for act in action:
        clone_state = copy.deepcopy(state)
        x1 = loc[0]
        y1 = loc[1]
        x2 = loc[0] + act[0]
        y2 = loc[1] + act[1]
        clone_state[x1][y1], clone_state[x2][y2] = clone_state[x2][y2], clone_state[x1][y1]
        new_node = copy.deepcopy(node)
        new_node.state = clone_state
        new_node.parent = node
        new_node.cost = node.cost + 1
        if (act[0] == 1):
            new_node.action = "DOWN"
        if (act[0] == -1):
            new_node.action = "UP"
        if (act[1] == -1):
            new_node.action = "LEFT"
        if (act[1] == 1):
            new_node.action = "RIGHT"
        child.append(new_node)
    return child
for child in expand(a):
    child.print_node()
```

```python
def BFS(problem):
    node = Node(problem.initial, "", "", 0)
    if problem.is_goal(node.state) == True:
        return node
    frontier = Frontier()
    frontier.enqueue(node)
    reached = [node.state]
    while frontier.is_empty() != True:
        node = frontier.dequeue()
        for child in expand(node):
            state = child.state
            if problem.is_goal(state) == True:
                return child
            if state not in reached and frontier.is_contain(child) == False:
                reached.append(state)
                frontier.enqueue(child)
    failure = "Khong the giai tro choi"
    return failure
```

```python
import tkinter as tk
import time
def perform(state, cells_state):
    for i in range(len(state)):
        for j in range(len(state[0])):
            if state[i][j] == 0:
                cells_state[i][j].config(text="", bg="#7d7d7d")
            else:
                cells_state[i][j].config(text=str(state[i][j]), bg="#f0f0f0")
def handle_click(lbl_status, btn_run, problem, root, cells_state):
    lbl_status.config(text="Đang tính toán ...", fg="black")
    btn_run.config(state=tk.DISABLED)
    root.update()
    result = BFS(problem)
    if type(result) is str:
        lbl_status.config(text=result, fg="red")
        root.update()
        return
    current = result
    path = []
    while current != "":
        path.append(current)
        current = current.parent
    path.reverse()
    for i, node in enumerate(path):
        perform(node.state, cells_state)
        if (i==0):
            lbl_status.config(text="Trạng thái ban đầu", fg="black")
        else:
            lbl_status.config(text=f"Bước {i}, hành động thực hiện là {node.action}", fg="black")
        root.update()
        time.sleep(1.25)
    lbl_status.config(text=f"Đã thực hiện xong, tổng số bước là {len(path)-1}", fg="green")
def main() :
    initial = [
        [8, 2, 3],
        [1, 6, 4],
        [7, 0, 5]
    ]
    goal = [
        [1, 2, 3],
        [4, 5, 6],
        [7, 8, 0]
    ]
    problem = Problem(initial, goal)
    cells_state = [[None, None, None],
                   [None, None, None],
                   [None, None, None]]
    root = tk.Tk()
    root.title("8 Puzzle BFS Type 2")
    for i in range(3):
        for j in range(3):
            lbl = tk.Label(root, text="", font=("Time New Roman", 20),width=5, height=2, relief="solid", bd=1)
            lbl.grid(row=i, column=j, padx=2, pady=2)
            cells_state[i][j] = lbl
    perform(problem.initial, cells_state)
    lbl_status = tk.Label(root, text="Bấm nút để giải", font=("Time New Roman", 10))
    lbl_status.grid(row=3, column=0,columnspan=3, pady=2)
    btn_run = tk.Button(root, text="GIẢI", font=("Time New Roman", 14), bg="#16f646", fg="#ffffff", command= lambda: handle_click(lbl_status,btn_run, problem, root, cells_state))
    btn_run.grid(row=4, column=0, columnspan=3, pady=2)
    root.mainloop()
main()

```
