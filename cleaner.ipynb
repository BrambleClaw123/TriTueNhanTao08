```python
def INTERPRET_INPUT(percept):
    for i in range(len(percept)):
        for j in range(len(percept[0])):
            if percept[i][j] == 'X':
                state = {
                    "top" : -1,
                    "bot" : -1,
                    "left" : -1,
                    "right" : -1,
                    "loc" : (i, j)
                } #0 la sach, 1 la do, -1 la k the di
                if (i!=0):
                    state["top"] = 1 if percept[i-1][j] == 1 else 0
                if (i!=len(percept)-1):
                    state["bot"] = 1 if percept[i+1][j] == 1 else 0
                if (j!=0):
                    state["left"] = 1 if percept[i][j-1] == 1 else 0
                if (j!=len(percept[0])-1):
                    state["right"] = 1 if percept[i][j+1] == 1 else 0
                return state
percept = [[0, 1, 1],
           [1, 1, 0],
           [1, 1, 'X']]
print(INTERPRET_INPUT(percept))
```

```python
def rules(state):
    rule = []
    for key, value in state.items():
        if key == "loc":
            continue
        if value == 1 and key == "top":
            rule.append((-1, 0))
        elif value == 1 and key == "bot":
            rule.append((1, 0))
        elif value == 1 and key == "left":
            rule.append((0, -1))
        elif value == 1 and key == "right":
            rule.append((0, 1))
    if len(rule) != 0:
        return rule
    else:
        for key, value in state.items():
            if key == "loc":
                continue
            if value == 0 and key == "top":
                rule.append((-1, 0))
            elif value == 0 and key == "bot":
                rule.append((1, 0))
            elif value == 0 and key == "left":
                rule.append((0, -1))
            elif value == 0 and key == "right":
                rule.append((0, 1))
        return rule
state = INTERPRET_INPUT(percept)
print(rules(state))
```

```python
def RULE_MATCH(state):
    rule = rules(state)
    return rule
print(RULE_MATCH(state))
```

```python
import random
```

```python
def ACTION(rule, percept, state):
    action = random.choice(rule)
    action_x = action[0]
    action_y = action[1]
    new_state = (state["loc"][0] + action_x, state["loc"][1] + action_y)
    if (action_x == 1):
        clear(state["loc"][0], state["loc"][1], new_state[0], new_state[1], percept)
        print('Robot di chuyen xuong')
    if (action_x == -1):
        clear(state["loc"][0], state["loc"][1],new_state[0], new_state[1], percept)
        print('Robot di chuyen len')
    if (action_y == -1):
        clear(state["loc"][0], state["loc"][1], new_state[0], new_state[1], percept)
        print('Robot di chuyen qua trai')
    if (action_y == 1):
        clear(state["loc"][0], state["loc"][1], new_state[0], new_state[1], percept)
        print('Robot di chuyen qua phai')

def clear(x1, y1, x2, y2, matrix) :
    matrix[x1][y1] = 0
    matrix[x2][y2] = "X"

def print_mat(matrix) :
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            print(matrix[i][j], end=" ")
        print()

rule = RULE_MATCH(state)
ACTION(rule, percept, state)
print_mat(percept)
```

```python
def random_percept(numr, numc):
    ran1 = [0, 1, "X"]
    ran2 = [0, 1]
    has_robot = False
    percept = []
    for i in range(numr):
        row = []
        for j in range(numc):
            if (has_robot == False):
                val = random.choice(ran1)
                row.append(val)
                if val == "X":
                    has_robot = True
            else:
                val = random.choice(ran2)
                row.append(val)
        percept.append(row)
    return percept

def check(percept):
    for row in percept:
        for value in row:
            if (value == 'X'):
                continue
            if (value == 1):
                return False
    return True

def main() :
    time = 50000
    percept = random_percept(3,3)
    print("San nha luc dau")
    print_mat(percept)
    if (check(percept) == True):
        print("San da duoc lau xong")
        return
    while (time != 0) :
        state = INTERPRET_INPUT(percept)
        rule = RULE_MATCH(state)
        ACTION(rule, percept, state)
        print_mat(percept)
        c = check(percept)
        if (c == True):
            print("San da duoc lau xong")
            return
        else:
            time -= 1
    print("Lau san that bai")
    
main()
```
