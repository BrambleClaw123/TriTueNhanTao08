```python
class Problem:
    def __init__(self, initial, goal):
        self.initial = initial
        self.goal = goal
    def is_goal(self, state):
        for i in range(len(state)):
            for j in range(len(state[0])):
                if self.goal[i][j] != state[i][j]:
                    return False
        return True
problem = Problem([[1, 9, 8],[0, 5, 6], [7,2,3]], [[1, 2, 3],[4, 5, 6], [7,8,0]])
print(problem.is_goal(a.state))
```

```python
class Frontier:
    def __init__(self):
        self.queue = []
    def enqueue(self, node):
        self.queue.append(node)
    def dequeue(self):
        return self.queue.pop(0)
    def is_empty(self):
        return True if len(self.queue) == 0 else False
    def is_contain(self, node):
        return True if node in self.queue else False
frontier = Frontier()
frontier.enqueue(a)
print(frontier.queue[0].state)
print(frontier.is_empty())
print(frontier.is_contain(a))
b = frontier.dequeue()
print(b.state)
print(frontier.is_empty())
```

```python
import copy
def expand_step_cost(node):
    state = node.state
    action = []
    for i in range(len(state)):
        for j in range(len(state[0])):
            if (state[i][j] == 0):
                loc = (i, j)
                if (j!=0):
                    action.append((0, -1))
                if (j!= len(state[0])-1):
                    action.append((0, 1))
                if (i!=0):
                    action.append((-1, 0))
                if (i!=len(state)-1):
                    action.append((1, 0))
    child = []
    for act in action:
        clone_state = copy.deepcopy(state)
        x1 = loc[0]
        y1 = loc[1]
        x2 = loc[0] + act[0]
        y2 = loc[1] + act[1]
        clone_state[x1][y1], clone_state[x2][y2] = clone_state[x2][y2], clone_state[x1][y1]
        new_node = Node(clone_state, node, "", node.cost + 1)
        if (act[0] == 1):
            new_node.action = "DOWN"
        if (act[0] == -1):
            new_node.action = "UP"
        if (act[1] == -1):
            new_node.action = "LEFT"
        if (act[1] == 1):
            new_node.action = "RIGHT"
        child.append(new_node)
    return child
for child in expand_step_cost(a):
    child.print_node()
```

```python
def BFS_type_1(problem):
    node = Node(problem.initial, "", "", 0)
    frontier = Frontier()
    frontier.enqueue(node)
    reached = [node.state]
    while frontier.is_empty() == False:
        node = frontier.dequeue()
        if problem.is_goal(node.state) == True:
            return node
        for child in expand_step_cost(node):
            state = child.state
            if state not in reached and frontier.is_contain(child) == False:
                reached.append(state)
                frontier.enqueue(child)
    failure = "Không thể giải được"
    return failure
    
```

```python
def BFS_type_2(problem):
    node = Node(problem.initial, "", "", 0)
    if problem.is_goal(node.state) == True:
        return node
    frontier = Frontier()
    frontier.enqueue(node)
    reached = [node.state]
    while frontier.is_empty() != True:
        node = frontier.dequeue()
        for child in expand_step_cost(node):
            state = child.state
            if problem.is_goal(state) == True:
                return child
            if state not in reached and frontier.is_contain(child) == False:
                reached.append(state)
                frontier.enqueue(child)
    failure = "Không thể giải được"
    return failure
```

```python
def IDS(problem):
    def DLS(problem, depth):
        node = Node(problem.initial, "", "", 0)
        if problem.is_goal(node.state) == True:
            return node
        result = "Không thể giải được"
        frontier = Frontier_2()
        frontier.push(node)
        reached = [node.state]
        while frontier.is_empty() != True:
            node = frontier.pop()
            if node.cost >= depth:
                result = "cutoff"
                continue
            for child in expand_step_cost(node):
                state = child.state
                if problem.is_goal(state):
                    return child
                if state not in reached and not frontier.is_contain(child):
                    reached.append(state)
                    frontier.push(child)
        return result
            
    depth = 0
    while True:
        result = DLS(problem, depth)
        if result != "cutoff":
            return result
        depth += 1
```

```python
def manhattan(state):
    goal = [[1, 2, 3],
            [4, 5, 6],
            [7, 8, 0]]
    cost = 0
    for i in range(3):
        for j in range(3):
            if state[i][j] == 1:
                cost += abs(i - 0) + abs(j - 0)
            elif state[i][j] == 2:
                cost += abs(i - 0) + abs(j - 1)
            elif state[i][j] == 3:
                cost += abs(i - 0) + abs(j - 2)
            elif state[i][j] == 4:
                cost += abs(i - 1) + abs(j - 0)
            elif state[i][j] == 5:
                cost += abs(i - 1) + abs(j - 1)
            elif state[i][j] == 6:
                cost += abs(i - 1) + abs(j - 2)
            elif state[i][j] == 7:
                cost += abs(i - 2) + abs(j - 0)
            elif state[i][j] == 8:
                cost += abs(i - 2) + abs(j - 1)
    return cost
print(manhattan([[1, 2, 3],[4, 0, 6], [7, 5, 8]]))
```

```python
def SHC(problem):
    current_state = problem.initial
    current_node = Node(current_state, "", "", manhattan(current_state))
    if (problem.is_goal(current_state)):
        return current_state
    while True:
        local_maximum = True
        for child in expand_hn(current_node):
            if child.cost < current_node.cost :
                current_node = child
                if problem.is_goal(current_node.state):
                    return current_node
                local_maximum = False
            break
        if local_maximum:
            break
    return "Chỉ đạt trạng thái tối ưu cục bộ, không thể giải đến goal."    
```

```python
def LBS(problem, k = 4):
    node = Node(problem.initial, "", "", manhattan(problem.initial))
    if problem.is_goal(node.state):
        return node
    current_node_set = [child for child in expand_hn(node)]
    reached = []
    while True:
        neighboors = []
        for n in current_node_set:
            for child_n in expand_hn(n):
                if child_n not in neighboors and child_n.state not in reached:
                    neighboors.append(child_n)
                    reached.append(child_n.state)
        if len(neighboors) == 0:
            current_node_set.sort()
            return current_node_set[0]
        for neighboor in neighboors:
            if problem.is_goal(neighboor.state):
                return neighboor
        neighboors.sort()
        if len(neighboors) <= k:
            current_node_set = neighboors.copy()
        else:
            current_node_set = neighboors[0:k]
```

```python
def SAA(problem):
    current_node = Node(problem.initial, "", "", manhattan(problem.initial))
    T = 1000
    Tmin = 0.001
    alpha = 0.99
    while T > Tmin:
        if problem.is_goal(current_node.state):
            return current_node
        next_state = random_neighboors(current_node)
        denta = next_state.cost - current_node.cost
        if denta < 0:
            current_node = next_state
        else:
            p = math.exp(-denta/T)
            if random.random() < p:
                current_node = next_state
        T = T*alpha
    return "Tìm kiếm thất bại."
```

```python
def is_goal_blind_start(states, goal):
    for i in range(3):
        for j in range(3):
            if states[0][i][j] != goal[i][j] or states[1][i][j] != goal[i][j]:
                return False
    return True
```

```python
def blind_start_search():
    belief_start = [[[1, 2, 3],
                     [4, 0, 6],
                     [7, 5, 8]],
                    [[0, 1, 3],
                     [4, 2, 6],
                     [7, 5, 8]]]
    goal = [[1, 2, 3],
            [4, 5, 6],
            [7, 8, 0]]
    # Chọn thuật toán Greedy Search Algorithm
    node = BeliefNode(belief_start, "", "", (manhattan(belief_start[0])+ manhattan(belief_start[1]))/2)
    frontier = []
    reached = []
    heapq.heappush(frontier, node)
    while len(frontier) > 0:
        node = heapq.heappop(frontier)
        if is_goal_blind_start(node.states, goal):
            return node
        reached.append(node.states)
        for child in expand_belief(node):
            if child not in frontier and child.states not in reached:
                heapq.heappush(frontier, child)
    return "Tìm kiếm thất bại"
```

```python
def move(action, state, loc):
    x1 = loc[0]
    y1 = loc[1]
    x2 = loc[0] + action[0]
    y2 = loc[1] + action[1]
    x3 = loc[0] + 1 if loc[0] != 2 and action[0] != 1 else loc[0]
    y3 = loc[1] + 1 if x3 != loc[0] and loc[1] != 2 and action[1] != 1 else loc[1]
    result_1 = copy.deepcopy(state)
    result_2 = copy.deepcopy(state)
    result_1[x1][y1], result_1[x2][y2] = result_1[x2][y2], result_1[x1][y1]
    result_2[x1][y1], result_2[x3][y3] = result_2[x3][y3], result_2[x1][y1]

    results = [result_1]
    if result_2 != result_1 and result_2 != state:
        results.append(result_2)

    return results
```

```python
def or_search(state, problem, path):
    if problem.is_goal(state):
        return []
    if state in path:
        return "failure"
    for action in get_actions(state)[1]:
        result_states = move(action, state, get_actions(state)[0])
        plan = and_search(result_states, problem, path + [state])
        if plan != "failure":
            return {action : plan}
    return "failure"
        
```

```python
def and_search(states, problem, path):
    plan = {}
    for state in states:
        plan_s = or_search(state, problem, path)
        if plan_s == "failure":
            return "failure"
        state_tuple = tuple(tuple(row) for row in state)
        plan[state_tuple] = plan_s
    return plan
```

```python
def backtracking(problem, node, path):
    if problem.is_goal(node.state):
        return node
    for child in expand_gn(node):
        path.append(child)
        result = backtracking(problem, child, path)
        if result != "Không thể giải":
            return result
        path.pop()
    return "Không thể giải"
```

```python
def foward_checking(problem, node, path):
    if problem.is_goal(node.state):
        return node
    for child in expand_gn(node):
        if child.state not in path:
            path.append(child)
            result = foward_checking(problem, child, path)
            if result != "Không thể giải":
                return result
            path.pop()
    return "Không thể giải"
```

```python
import tkinter as tk
from tkinter import ttk
import time
import random

def perform(state, cells_state):
    for i in range(len(state)):
        for j in range(len(state[0])):
            val = state[i][j]
            if val == 0:
                cells_state[i][j].config(text="", bg="#D9C3B0", relief="sunken", bd=3)
            else:
                cells_state[i][j].config(text=str(val), bg="#F5E6D3", fg="#4A2E1B", relief="raised", bd=4)

def write_to_log(txt_log, message):
    txt_log.config(state=tk.NORMAL)       
    txt_log.insert(tk.END, message + "\n") 
    txt_log.config(state=tk.DISABLED)     
    txt_log.see(tk.END)                   

def format_state(state):
    return "\n".join(" ".join(str(val) for val in row) for row in state)

def open_bsa_window(parent_root, selected_algo):
    algo_name = selected_algo.get()
    bsa_win = tk.Toplevel(parent_root)
    bsa_win.title(f"Simulation: {algo_name}")
    bsa_win.configure(bg="#EFE5D3")
    bsa_win.resizable(False, False)

    def on_closing_popup():
        selected_algo.set("BFS Type 1")
        bsa_win.destroy()
        
    bsa_win.protocol("WM_DELETE_WINDOW", on_closing_popup)
    
    if algo_name == "Partial Start":
        belief_start = [[[1, 2, 3], [4, 5, 6], [0, 7, 8]], [[1, 2, 3], [4, 5, 6], [8, 0, 7]]]
        partial_state = [[1, 2, 0], [4, 0, 0], [0, 8, 0]]
    else: 
        belief_start = [[[1, 2, 3], [4, 0, 6], [7, 5, 8]], [[0, 1, 3], [4, 2, 6], [7, 5, 8]]]
        partial_state = None
    
    top_panel = tk.Frame(bsa_win, bg="#E6D5C3", padx=15, pady=10, bd=2, relief="groove")
    top_panel.grid(row=0, column=0, columnspan=2, sticky="ew", padx=15, pady=(15, 5))
    
    btn_run_bsa = tk.Button(top_panel, text="PLAY", font=("Times New Roman", 10, "bold"), bg="#F5E6D3", fg="#2E5A1C", relief="raised", bd=3, width=8, pady=2)
    btn_run_bsa.pack(side=tk.LEFT, padx=5)

    btn_reset_bsa = tk.Button(top_panel, text="RESET", font=("Times New Roman", 10, "bold"), bg="#F5E6D3", fg="#A52A2A", relief="raised", bd=3, width=8, pady=2)
    btn_reset_bsa.pack(side=tk.LEFT, padx=5)
    
    tk.Label(top_panel, text=f"Thuật toán: {algo_name}", font=("Times New Roman", 12, "bold"), bg="#E6D5C3", fg="#4A2E1B").pack(side=tk.LEFT, padx=(25, 5))

    if partial_state:
        mini_frame = tk.Frame(top_panel, bg="#D9C3B0", bd=1, relief="solid", padx=2, pady=2)
        mini_frame.pack(side=tk.RIGHT, padx=15)
        tk.Label(mini_frame, text="Partial", font=("Times New Roman", 7, "bold"), bg="#D9C3B0", fg="#4A2E1B").grid(row=0, column=0, columnspan=3)
        for i in range(3):
            for j in range(3):
                val = partial_state[i][j]
                t = "" if val == 0 else str(val)
                bg_c = "#D9C3B0" if val == 0 else "#F5E6D3"
                tk.Label(mini_frame, text=t, font=("Times New Roman", 8, "bold"), bg=bg_c, fg="#4A2E1B", width=2, height=1, relief="ridge").grid(row=i+1, column=j)

    matrices_frame = tk.Frame(bsa_win, bg="#EFE5D3")
    matrices_frame.grid(row=1, column=0, padx=(15, 5), pady=(5, 15), sticky="n")
    
    matrix1_panel = tk.Frame(matrices_frame, bg="#D9C3B0", padx=10, pady=10, bd=3, relief="sunken")
    matrix1_panel.pack(side=tk.TOP, pady=(0, 10))
    tk.Label(matrix1_panel, text="State 1", bg="#D9C3B0", font=("Times New Roman", 10, "bold"), fg="#4A2E1B").grid(row=0, column=0, columnspan=3)
    
    matrix2_panel = tk.Frame(matrices_frame, bg="#D9C3B0", padx=10, pady=10, bd=3, relief="sunken")
    matrix2_panel.pack(side=tk.TOP)
    tk.Label(matrix2_panel, text="State 2", bg="#D9C3B0", font=("Times New Roman", 10, "bold"), fg="#4A2E1B").grid(row=0, column=0, columnspan=3)
    
    cells_1 = [[None]*3 for _ in range(3)]
    cells_2 = [[None]*3 for _ in range(3)]
    
    for i in range(3):
        for j in range(3):
            lbl1 = tk.Label(matrix1_panel, text="", font=("Times New Roman", 20, "bold"), width=3, height=1)
            lbl1.grid(row=i+1, column=j, padx=4, pady=4)
            cells_1[i][j] = lbl1
            
            lbl2 = tk.Label(matrix2_panel, text="", font=("Times New Roman", 20, "bold"), width=3, height=1)
            lbl2.grid(row=i+1, column=j, padx=4, pady=4)
            cells_2[i][j] = lbl2

    perform(belief_start[0], cells_1)
    perform(belief_start[1], cells_2)

    log_panel = tk.Frame(bsa_win, bg="#EFE5D3")
    log_panel.grid(row=1, column=1, padx=(5, 15), pady=(5, 15), sticky="nsew")
    txt_log = tk.Text(log_panel, width=40, height=22, font=("Times New Roman", 10, "bold"), bg="#FDFBF7", fg="#3E2723", relief="sunken", bd=2, padx=8, pady=8)
    txt_log.pack(fill=tk.BOTH, expand=True)
    txt_log.config(state=tk.DISABLED)
    
    lbl_status = tk.Label(bsa_win, text=f"Sẵn sàng chạy mô phỏng {algo_name}", font=("Times New Roman", 11, "italic"), bg="#EFE5D3", fg="#4A2E1B", wraplength=400)
    lbl_status.grid(row=2, column=0, columnspan=2, pady=(0, 15), padx=15)
    
    def handle_bsa_play():
        btn_run_bsa.config(state=tk.DISABLED)
        lbl_status.config(text=f"Đang tính toán thuật toán {algo_name}...", fg="#8B4513")
        
        txt_log.config(state=tk.NORMAL)
        txt_log.delete("1.0", tk.END)
        txt_log.config(state=tk.DISABLED)
        bsa_win.update()
        
        try:
            if algo_name == "Partial Start": result = partial_start_search()
            else: result = blind_start_search() 
        except Exception as e:
            result = f"Lỗi gọi hàm: {e}"
        
        if type(result) is str: 
            lbl_status.config(text=result, fg="#A52A2A")
            write_to_log(txt_log, f"Lỗi/Kết quả: {result}")
            btn_run_bsa.config(state=tk.NORMAL)
            return
            
        current = result
        path = []
        while current and current != "":
            path.append(current)
            if hasattr(current, 'parent') and current.parent != "": current = current.parent
            else: break
        path.reverse()
        
        for i, node in enumerate(path):
            perform(node.states[0], cells_1)
            perform(node.states[1], cells_2)
            
            action_name = getattr(node, 'action', '') 
            if i == 0: mess = "Trạng thái ban đầu"
            else: mess = f"Bước {i}: Hành động thực hiện [{action_name}]"
                
            lbl_status.config(text=mess, fg="#4A2E1B")
            mess += f"\n[State 1]:\n{format_state(node.states[0])}\n[State 2]:\n{format_state(node.states[1])}\n"
            
            write_to_log(txt_log, mess)
            bsa_win.update()
            time.sleep(0.75) 
            
        final_mess = f"Đã xong, tổng số bước là {len(path)-1}"
        write_to_log(txt_log, f"\n========\n{final_mess}")
        
        actions_str = " -> ".join(getattr(n, 'action', '') for n in path if getattr(n, 'action', '') != "")
        lbl_status.config(text=f"{final_mess}\nCác bước: {actions_str}", fg="#2E5A1C", wraplength=400)
        btn_run_bsa.config(state=tk.NORMAL)

    def handle_bsa_reset():
        perform(belief_start[0], cells_1)
        perform(belief_start[1], cells_2)
        txt_log.config(state=tk.NORMAL)
        txt_log.delete("1.0", tk.END)
        txt_log.config(state=tk.DISABLED)
        lbl_status.config(text=f"Sẵn sàng chạy mô phỏng {algo_name}", fg="#4A2E1B")

    btn_run_bsa.config(command=handle_bsa_play)
    btn_reset_bsa.config(command=handle_bsa_reset)

def handle_click(lbl_status, btn_run, problem, root, cells_state, txt_log, selected_algo):
    lbl_status.config(text="Đang tính toán ...", fg="#8B4513") 
    btn_run.config(state=tk.DISABLED)
    txt_log.config(state=tk.NORMAL)
    txt_log.delete("1.0", tk.END)
    txt_log.config(state=tk.DISABLED)
    root.update()
    
    algo_name = selected_algo.get()
    write_to_log(txt_log, f"Đang chạy thuật toán {algo_name}")
    root.update()
    
    result = ""
    try:
        if algo_name == "Backtracking":
            start_node = Node(problem.initial, "", "", 0)
            result = backtracking(problem, start_node, [problem.initial])
        elif algo_name == "Foward Checking":
            start_node = Node(problem.initial, "", "", 0)
            result = foward_checking(problem, start_node, [problem.initial])
            
        elif algo_name == "BFS Type 1": result = BFS_type_1(problem)
        elif algo_name == "BFS Type 2": result = BFS_type_2(problem)
        elif algo_name == "DFS": result = DFS(problem)
        elif algo_name == "IDS": result = IDS(problem)
        elif algo_name == "UCS": result = UCS(problem)
        elif algo_name == "GSA": result = GSA(problem)
        elif algo_name == "A*": result = A_star(problem)
        elif algo_name == "IDA*": result = IDA_star(problem)
        elif algo_name == "SHC": result = SHC(problem)
        elif algo_name == "SAHC": result = SAHC(problem)
        elif algo_name == "RHC": result = RHC(problem)
        elif algo_name == "RRHC": result = RRHC(problem)
        elif algo_name == "LBS": result = LBS(problem)
        elif algo_name == "SAA": result = SAA(problem)
        elif algo_name == "Blind Goal": result = blind_goal_search(problem.initial)
        elif algo_name == "AND-OR Search":
            plan = and_or_graph_search(problem)
            if plan == "failure":
                lbl_status.config(text="Không tìm thấy kế hoạch khả thi!", fg="#A52A2A")
                write_to_log(txt_log, "Lỗi: Thuật toán trả về failure.")
                btn_run.config(state=tk.NORMAL)
                return
                
            write_to_log(txt_log, "Đã tính toán xong Cây Kế Hoạch AND-OR.\nBắt đầu giả lập môi trường ngẫu nhiên...\n")
            
            current_state = problem.initial
            current_plan = plan
            step = 1
            
            while current_plan and isinstance(current_plan, dict):
                perform(current_state, cells_state)
                root.update()
                time.sleep(0.75)
                
                action = list(current_plan.keys())[0]
                and_plan = current_plan[action]
                
                act_name = "RIGHT" if action == (0, 1) else "LEFT" if action == (0, -1) else "DOWN" if action == (1, 0) else "UP"
                
                mess = f"Bước {step}: Hành động [{act_name}]"
                lbl_status.config(text=mess, fg="#4A2E1B")
                write_to_log(txt_log, f"{mess}\nTrạng thái hiện tại:\n{format_state(current_state)}\n")
                
                loc, _ = get_actions(current_state)
                possible_outcomes = move(action, current_state, loc)
                actual_next_state = random.choice(possible_outcomes)
                
                write_to_log(txt_log, f" -> Môi trường phản hồi state nhánh: {possible_outcomes.index(actual_next_state) + 1}\n")
                
                if (hasattr(problem, 'goal') and actual_next_state == problem.goal) or actual_next_state == [[1, 2, 3], [4, 5, 6], [7, 8, 0]]:
                    current_state = actual_next_state
                    break
                    
                state_key = tuple(tuple(row) for row in actual_next_state)
                current_plan = and_plan.get(state_key, [])
                current_state = actual_next_state
                step += 1
                
            perform(current_state, cells_state)
            lbl_status.config(text="Đã tới Đích bằng Kế hoạch AND-OR!", fg="#2E5A1C")
            write_to_log(txt_log, "\n========\nĐÃ ĐẾN ĐÍCH AN TOÀN!")
            problem.initial = current_state
            btn_run.config(state=tk.NORMAL)
            return 
            
    except Exception as e:
        result = f"Lỗi chạy thuật toán: {e}"
        
    if type(result) is str:
        lbl_status.config(text=result, fg="#A52A2A")
        write_to_log(txt_log, f"Lỗi/Kết quả: {result}")
        btn_run.config(state=tk.NORMAL)
        root.update()
        return
        
    current = result
    path = []
    while current != "":
        path.append(current)
        if hasattr(current, 'parent') and current.parent != "": current = current.parent
        else: break
    path.reverse()

    for i, node in enumerate(path):
        state_to_draw = node.state if hasattr(node, 'state') else node
        perform(state_to_draw, cells_state)
        
        # Phiên dịch hành động của CSP từ tuple sang chữ cái
        action_name = getattr(node, 'action', '')
        if isinstance(action_name, tuple):
            action_name = "RIGHT" if action_name == (0, 1) else "LEFT" if action_name == (0, -1) else "DOWN" if action_name == (1, 0) else "UP"
        
        if i == 0: mess = "Trạng thái ban đầu"
        else: mess = f"Bước {i}: Hành động thực hiện [{action_name}]"
            
        lbl_status.config(text=mess, fg="#4A2E1B")
        mess += "\nTrạng thái: \n"
        for row in state_to_draw:
            for value in row: mess += str(value) + " "
            mess += "\n"
        mess += "\n"
            
        write_to_log(txt_log, mess)
        root.update()
        time.sleep(0.75)
        
    final_mess = f"Đã xong, tổng số bước là {len(path)-1}"
    write_to_log(txt_log, f"\n========\n{final_mess}")

    actions_str = []
    for n in path:
        act = getattr(n, 'action', '')
        if isinstance(act, tuple):
            act = "RIGHT" if act == (0, 1) else "LEFT" if act == (0, -1) else "DOWN" if act == (1, 0) else "UP"
        if act != "": actions_str.append(act)
        
    final_mess += f"\nCác bước đã thực hiện\n{' -> '.join(actions_str)}"
    lbl_status.config(text=final_mess, fg="#2E5A1C", wraplength=320)
    
    problem.initial = state_to_draw
    btn_run.config(state=tk.NORMAL)

def handle_reset(problem, cells_state, lbl_status, txt_log, original_initial):
    problem.initial = [row[:] for row in original_initial]
    perform(problem.initial, cells_state)
    txt_log.config(state=tk.NORMAL)
    txt_log.delete("1.0", tk.END)
    txt_log.config(state=tk.DISABLED)
    lbl_status.config(text="Sẵn sàng chạy mô phỏng", fg="#4A2E1B")

def main():
    initial = [[0, 4, 3],
               [2, 5, 6],
               [1, 7, 8]]
    goal    = [[1, 2, 3],
               [4, 5, 6],
               [7, 8, 0]]
               
    problem = Problem(initial, goal)
        
    cells_state = [[None]*3 for _ in range(3)]
    
    root = tk.Tk()
    root.title("8 Puzzle Simulation")
    
    BG_MAIN = "#EFE5D3"      
    BG_PANEL = "#E6D5C3"     
    COLOR_TEXT = "#4A2E1B"   
    
    root.configure(bg=BG_MAIN)
    root.resizable(False, False) 
    
    style = ttk.Style()
    style.theme_use('clam')
    style.configure("TCombobox", fieldbackground="#F5E6D3", background=BG_PANEL, 
                    foreground=COLOR_TEXT, darkcolor=COLOR_TEXT, lightcolor="#F5E6D3")
    
    top_panel = tk.Frame(root, bg=BG_PANEL, padx=15, pady=12, bd=2, relief="groove")
    top_panel.grid(row=0, column=0, columnspan=2, sticky="ew", padx=15, pady=(15, 5))

    btn_run = tk.Button(top_panel, text="PLAY", font=("Times New Roman", 10, "bold"), bg="#F5E6D3", fg="#2E5A1C", relief="raised", bd=3, width=8, pady=2)  
    btn_run.pack(side=tk.LEFT, padx=5)

    btn_reset = tk.Button(top_panel, text="RESET", font=("Times New Roman", 10, "bold"), bg="#F5E6D3", fg="#A52A2A", relief="raised", bd=3, width=8, pady=2) 
    btn_reset.pack(side=tk.LEFT, padx=5)

    lbl_algo = tk.Label(top_panel, text="Thuật toán:", font=("Times New Roman", 10, "italic"), bg=BG_PANEL, fg=COLOR_TEXT)
    lbl_algo.pack(side=tk.LEFT, padx=(25, 5))
    
    selected_algo = tk.StringVar(value="BFS Type 1")
    root.option_add('*TCombobox*Listbox.font', ("Times New Roman", 10, "bold"))
    algo_combobox = ttk.Combobox(top_panel, textvariable=selected_algo, state="readonly", width=16, font=("Times New Roman", 10, "bold"))
    
    # DANH SÁCH THÊM "CSP"
    algo_combobox['values'] = ("BFS Type 1", "BFS Type 2", "DFS", "IDS", "UCS", "GSA", "A*", "IDA*", "SHC", "SAHC", "RHC", "RRHC", "LBS", "SAA", "Blind Start", "Blind Goal", "Partial Start", "AND-OR Search", "Backtracking", "Foward Checking")
    algo_combobox.pack(side=tk.LEFT, padx=5)

    mini_goals_frame = tk.Frame(top_panel, bg=BG_PANEL)
    g1_frame = tk.Frame(mini_goals_frame, bg="#D9C3B0", bd=1, relief="solid", padx=2, pady=2)
    g1_frame.pack(side=tk.LEFT, padx=(15, 5))
    tk.Label(g1_frame, text="Goal 1", font=("Times New Roman", 7, "bold"), bg="#D9C3B0", fg=COLOR_TEXT).grid(row=0, column=0, columnspan=3)
    g2_frame = tk.Frame(mini_goals_frame, bg="#D9C3B0", bd=1, relief="solid", padx=2, pady=2)
    g2_frame.pack(side=tk.LEFT, padx=5)
    tk.Label(g2_frame, text="Goal 2", font=("Times New Roman", 7, "bold"), bg="#D9C3B0", fg=COLOR_TEXT).grid(row=0, column=0, columnspan=3)
    
    belief_goal = [[[1, 2, 3], [4, 5, 6], [7, 8, 0]], [[1, 2, 3], [8, 0, 4], [7, 6, 5]]]
    for i in range(3):
        for j in range(3):
            val1, val2 = belief_goal[0][i][j], belief_goal[1][i][j]
            bg1 = "#D9C3B0" if val1 == 0 else "#F5E6D3"
            bg2 = "#D9C3B0" if val2 == 0 else "#F5E6D3"
            tk.Label(g1_frame, text="" if val1 == 0 else str(val1), font=("Times New Roman", 8, "bold"), bg=bg1, fg=COLOR_TEXT, width=2, height=1, relief="ridge").grid(row=i+1, column=j)
            tk.Label(g2_frame, text="" if val2 == 0 else str(val2), font=("Times New Roman", 8, "bold"), bg=bg2, fg=COLOR_TEXT, width=2, height=1, relief="ridge").grid(row=i+1, column=j)

    def on_combobox_select(event):
        val = selected_algo.get()
        if val in ["Blind Start", "Partial Start"]:
            mini_goals_frame.pack_forget() 
            open_bsa_window(root, selected_algo)
        elif val == "Blind Goal":
            mini_goals_frame.pack(side=tk.LEFT)
        else:
            mini_goals_frame.pack_forget() 

    algo_combobox.bind("<<ComboboxSelected>>", on_combobox_select)
    
    matrix_panel = tk.Frame(root, bg="#D9C3B0", padx=15, pady=15, bd=3, relief="sunken")
    matrix_panel.grid(row=1, column=0, padx=(15, 5), pady=(5, 15), sticky="n")
    
    for i in range(3):
        for j in range(3):
            lbl = tk.Label(matrix_panel, text="", font=("Times New Roman", 24, "bold"), width=4, height=2)
            lbl.grid(row=i, column=j, padx=6, pady=6)
            cells_state[i][j] = lbl
            
    perform(problem.initial, cells_state)

    lbl_status = tk.Label(root, text="Sẵn sàng chạy mô phỏng", font=("Times New Roman", 11, "italic"), bg=BG_MAIN, fg=COLOR_TEXT, justify=tk.CENTER)
    lbl_status.grid(row=2, column=0, columnspan=2, pady=(0, 15), padx=15, sticky="ew")

    log_panel = tk.Frame(root, bg=BG_MAIN)
    log_panel.grid(row=1, column=1, padx=(5, 15), pady=(5, 5), sticky="nsew")

    txt_log = tk.Text(log_panel, width=38, height=15, font=("Times New Roman", 10, "bold"), bg="#FDFBF7", fg="#3E2723", relief="sunken", bd=2, padx=8, pady=8)
    txt_log.grid(row=0, column=0, sticky="nsew")
    log_panel.grid_rowconfigure(0, weight=1)
    log_panel.grid_columnconfigure(0, weight=1)
    txt_log.config(state=tk.DISABLED)

    btn_run.config(command=lambda: handle_click(lbl_status, btn_run, problem, root, cells_state, txt_log, selected_algo))
    btn_reset.config(command=lambda: handle_reset(problem, cells_state, lbl_status, txt_log, initial))

    root.mainloop()

if __name__ == "__main__":
    main()
```
